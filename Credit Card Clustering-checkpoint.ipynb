## Data Cleaning

```python
data.isnull().sum()
```

```python
# Drop row with missing CREDIT_LIMIT
data = data.dropna(subset=['CREDIT_LIMIT'])
# Replace missing MINIMUM_PAYMENTS with median
median_value = data['MINIMUM_PAYMENTS'].median()
data['MINIMUM_PAYMENTS']=data['MINIMUM_PAYMENTS'].fillna(median_value)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
# check and handle outlier
data_numeric=data.select_dtypes(exclude="object")
for col in data_numeric:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outlier=((data_numeric[col]>upper_bound)|(data_numeric[col]<lower_bound)).sum()
    total=data_numeric[col].shape[0]
    print(f"Total Outliers in {col} are {outlier}-{round(100*(outlier)/total,2)}%")
    if outlier>0:
        data=data.loc[(data[col]<=upper_bound) & (data[col]>=lower_bound)]
```

## Modeling

```python
sc = StandardScaler()
scaled_data = sc.fit_transform(data)
```

```python
num_of_clusters=[]
errors=[]
```

```python
for i in range(1,12):
    model=KMeans(n_clusters=i)
    model.fit(scaled_data)
    num_of_clusters.append(i)
    errors.append(model.inertia_)
```

```python
plt.plot(num_of_clusters, errors, marker='*')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (WCSS)")
plt.show()
```

```python
model = KMeans(n_clusters=5, random_state=42)
```

```python
model.fit(scaled_data)
```

```python
data["K_Means"] = model.labels_
```

```python
data
```

## Post-Clustering Analysis

```python
# Split data by cluster
df0 = data[data["K_Means"] == 0]
df1 = data[data["K_Means"] == 1]
df2 = data[data["K_Means"] == 2]
df3 = data[data["K_Means"] == 3]
df4 = data[data["K_Means"] == 4]
```

```python
# Plot each cluster
plt.scatter(df0["PURCHASES"], df0["CREDIT_LIMIT"], label="Cluster 0", s=60)
plt.scatter(df1["PURCHASES"], df1["CREDIT_LIMIT"], label="Cluster 1", s=60)
plt.scatter(df2["PURCHASES"], df2["CREDIT_LIMIT"], label="Cluster 2", s=60)
plt.scatter(df3["PURCHASES"], df3["CREDIT_LIMIT"], label="Cluster 3", s=60)
plt.scatter(df4["PURCHASES"], df4["CREDIT_LIMIT"], label="Cluster 4", s=60)

plt.xlabel("Total Purchases")
plt.ylabel("Credit Limit")
plt.title("Segmenting Customers Based on Spending and Credit Capacity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data['K_Means'].value_counts().plot.pie(autopct="%0.2f%%")
plt.title("Credit Card Distribution by Cluster")
plt.show()
```

```python
spending_features = ['PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'PURCHASES_TRX']
data.groupby('K_Means')[spending_features].mean().T.plot(kind='bar')
plt.title("Spending Behavior by Cluster")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data["TENURE_BIN"] = pd.cut(data["TENURE"], bins=[0, 6, 12], labels=["Short-Term", "Long-Term"])

sns.countplot(x="K_Means", hue="TENURE_BIN", data=data,color="pink")
plt.title("Cluster Distribution by Tenure")
plt.xlabel("Cluster Label")
plt.ylabel("Customer Count")
plt.legend(title="Tenure Category")
plt.tight_layout()
plt.show()
```

```python
data["PURCHASE_TYPE"] = np.where(data["ONEOFF_PURCHASES"] > data["INSTALLMENTS_PURCHASES"], "One-off", "Installments")

sns.countplot(x="K_Means", hue="PURCHASE_TYPE", data=data,color="pink")
plt.title("Cluster Distribution by Purchase Type")
plt.xlabel("Cluster Label")
plt.ylabel("Customer Count")
plt.legend(title="Purchase Type")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
data["PCA1"] = pca_result[:, 0]
data["PCA2"] = pca_result[:, 1]

sns.scatterplot(x="PCA1", y="PCA2", hue="K_Means", data=data, alpha=0.7)
plt.title("PCA Cluster Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```
